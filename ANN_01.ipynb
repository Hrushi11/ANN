# Q1) Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha * x, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x)

# Generate data
x = np.linspace(-10, 10, 100)

# Plots 
def plot_activation(x, value, label):
  # Plot activation functions
  plt.figure(figsize=(10, 6))
  plt.plot(x, value, label=label)
  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title(f'Activation Function: {label}')
  plt.legend()
  plt.show()
```

```python
# Plotting sigmoid
sigmoid_y = sigmoid(x)
plot_activation(x, sigmoid_y, label='sigmoid')
```

```python
# Plotting sigmoid
relu_y = relu(x)
plot_activation(x, relu_y, label='relu')
```

```python
# Plotting sigmoid
tanh_y = tanh(x)
plot_activation(x, tanh_y, label='tanh')
```

```python
# Plotting sigmoid
leaky_relu_y = leaky_relu(x)
plot_activation(x, leaky_relu_y, label='leaky_relu')
```

```python
# Plotting sigmoid
softmax_y = softmax(x)
plot_activation(x, softmax_y, label='softmax')
```
